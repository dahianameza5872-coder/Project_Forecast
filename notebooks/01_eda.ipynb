```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('../data/raw/online_retail_II.csv')

# Primeras filas
df.head()

# Información general
df.info()

# Nulos
df.isnull().sum()

# Descripción estadística
df.describe()

# Distribución de fechas si existe una columna de fecha
# Convierte a datetime y grafica
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.set_index('InvoiceDate').resample('ME').size().plot()
plt.title('Pedidos por mes')
plt.show()

# Histograma de ventas

df['Quantity'].hist(bins='auto', color='skyblue', edgecolor='black')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.title('Distribución de ventas')
plt.show()
```
